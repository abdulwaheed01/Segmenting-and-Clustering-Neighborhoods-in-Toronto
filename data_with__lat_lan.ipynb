# Add langitude and latitude to toronto data

### Geopatial Coordinates

```python
import pandas as pd
geospatial_cord = pd.read_csv('Geospatial_Coordinates.csv')
geospatial_cord
```

### toronto data without lat and long

```python
data = pd.read_csv('scrape_data.csv') 
data
```

### Make a dataframe of toronto data with lat and long

```python
df = pd.concat([data, geospatial_cord],join="outer",axis=1)

df.head()
```

```python
df.drop('Postal Code',axis=1,inplace=True)
```
